```python
import numpy as np
import matplotlib.pyplot as plt

from Option_Contract import OptionContract
from Binomial_Tree import crr_price

sp = float(input("Spot : "))
k = float(input("Strike : "))
sig = float(input("Volatility : "))
mat = float(input("Maturity : "))
rf = float(input("Risk Free Rate : "))
dvd = float(input("Dividends : "))
nb_simul = 100

spot_vect = np.arange(50,150,1)
call_prices_eur = [crr_price(OptionContract(opt_type="call", obs_type="european", spot=s, strike=k, sigma=sig, maturity=mat, risk_free=rf, div=dvd),nb_simul) for s in spot_vect]  
call_prices_us = [crr_price(OptionContract(opt_type="call", obs_type="american", spot=s, strike=k, sigma=sig, maturity=mat, risk_free=rf, div=dvd),nb_simul) for s in spot_vect]  
put_prices_eur = [crr_price(OptionContract(opt_type="put", obs_type="european", spot=s, strike=k, sigma=sig, maturity=mat, risk_free=rf, div=dvd),nb_simul) for s in spot_vect]  
put_prices_us = [crr_price(OptionContract(opt_type="put", obs_type="american", spot=s, strike=k, sigma=sig, maturity=mat, risk_free=rf, div=dvd),nb_simul) for s in spot_vect]  
call_prenium = np.array(call_prices_us) - np.array(call_prices_eur)
put_prenium = np.array(put_prices_us) - np.array(put_prices_eur)

fig, axes = plt.subplots(1,2,figsize=(18,10))
axes[0].plot(spot_vect, call_prenium, linewidth = 3, color = "red")
axes[0].axvline(x=sp, color = "black", linestyle = "--", linewidth = 1.5, label = "Current Spot")
axes[0].grid()
axes[0].legend()
axes[0].set_title("Call Prenium for US Options on Spot")
axes[0].set_xlabel("Spot")
axes[0].set_ylabel("Prenium")
axes[1].plot(spot_vect, put_prenium, linewidth = 3, color = "blue")
axes[1].axvline(x=sp, color = "black", linestyle = "--", linewidth = 1.5, label = "Current Spot")
axes[1].grid()
axes[1].legend()
axes[1].set_title("Put Prenium for US Options on Spot")
axes[1].set_xlabel("Spot")
axes[1].set_ylabel("Prenium")
plt.tight_layout()
plt.show()

vol_vect = np.arange(0.05,1,0.01)
call_prices_eur = [crr_price(OptionContract(opt_type="call", obs_type="european", spot=sp, strike=k, sigma=s, maturity=mat, risk_free=rf, div=dvd),nb_simul) for s in vol_vect]  
call_prices_us = [crr_price(OptionContract(opt_type="call", obs_type="american", spot=sp, strike=k, sigma=s, maturity=mat, risk_free=rf, div=dvd),nb_simul) for s in vol_vect]  
put_prices_eur = [crr_price(OptionContract(opt_type="put", obs_type="european", spot=sp, strike=k, sigma=s, maturity=mat, risk_free=rf, div=dvd),nb_simul) for s in vol_vect]  
put_prices_us = [crr_price(OptionContract(opt_type="put", obs_type="american", spot=sp, strike=k, sigma=s, maturity=mat, risk_free=rf, div=dvd),nb_simul) for s in vol_vect]  
call_prenium = np.array(call_prices_us) - np.array(call_prices_eur)
put_prenium = np.array(put_prices_us) - np.array(put_prices_eur)

fig, axes = plt.subplots(1,2,figsize=(18,10))
axes[0].plot(vol_vect, call_prenium, linewidth = 3, color = "red")
axes[0].axvline(x=sig, color = "black", linestyle = "--", linewidth = 1.5, label = "Current Vol")
axes[0].grid()
axes[0].legend()
axes[0].set_title("Call Prenium for US Options on Volatility")
axes[0].set_xlabel("Volatility")
axes[1].set_ylabel("Prenium")
axes[1].plot(vol_vect, put_prenium, linewidth = 3, color = "blue")
axes[1].axvline(x=sig, color = "black", linestyle = "--", linewidth = 1.5, label = "Current Vol")
axes[1].grid()
axes[1].legend()
axes[1].set_title("Put Prenium for US Options on Volatility")
axes[1].set_xlabel("Volatility")
axes[1].set_ylabel("Prenium")
plt.tight_layout()
plt.show()
```
