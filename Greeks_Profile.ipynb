```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from Option_Contract import OptionContract
from Black_Scholes import greeks_bs

k = float(input("Strike : "))
sig = float(input("Volatility : "))
mat = float(input("Maturity : "))
rf = float(input("Risk Free Rate : "))
dvd = float(input("Dividends : "))

spot_vect = np.arange(1,150,1)
contract_call = [OptionContract(opt_type="call", obs_type="european", spot=s, strike=k, sigma=sig, maturity=mat, risk_free=rf, div=dvd) for s in spot_vect]
contract_put = [OptionContract(opt_type="put", obs_type="european", spot=s, strike=k, sigma=sig, maturity=mat, risk_free=rf, div=dvd) for s in spot_vect]

delta_call =[greeks_bs(c).loc["Delta","Value"] for c in contract_call]
delta_put = [greeks_bs(c).loc["Delta","Value"] for c in contract_put]
fig, ax = plt.subplots(figsize = (12,7))
ax.plot(spot_vect, delta_call, color = "blue", linewidth = 3, label = "Delta of a Call Option")
ax.plot(spot_vect, delta_put, color = "orange", linewidth = 3, label = "Delta of a Put Option")
ax.set_xlabel("Spot")
ax.set_ylabel("Delta")
ax.grid()
ax.legend()
plt.show()

fig, ax = plt.subplots(figsize=(12,7))
gamma_call =[greeks_bs(c).loc["Gamma","Value"] for c in contract_call]
gamma_put = [greeks_bs(c).loc["Gamma","Value"] for c in contract_put]
ax.plot(spot_vect, gamma_call, color = "blue", linewidth = 3, linestyle = "--", label = "Gamma of a Call Option")
ax.plot(spot_vect, gamma_put, color = "orange", linewidth = 3, alpha = 0.6, label = "Gamma of a Put Option")
ax.set_xlabel("Spot")
ax.set_ylabel("Gamma")
ax.grid()
ax.legend()
plt.show()

fig, ax = plt.subplots(figsize=(12,7))
vega_call =[greeks_bs(c).loc["Vega","Value"] for c in contract_call]
vega_put = [greeks_bs(c).loc["Vega","Value"] for c in contract_put]
ax.plot(spot_vect, vega_call, color = "blue", linewidth = 3, linestyle = "--", label = "Vega of a Call Option")
ax.plot(spot_vect, vega_put, color = "orange", linewidth = 3, alpha = 0.6, label = "Vega of a Put Option")
ax.set_xlabel("Spot")
ax.set_ylabel("Vega")
ax.grid()
ax.legend()
plt.show()

fig, ax = plt.subplots(figsize=(12,7))
theta_call =[greeks_bs(c).loc["Theta","Value"] for c in contract_call]
theta_put = [greeks_bs(c).loc["Theta","Value"] for c in contract_put]
ax.plot(spot_vect, theta_call, color = "blue", linewidth = 3, linestyle = "--", label = "Theta of a Call Option")
ax.plot(spot_vect, theta_put, color = "orange", linewidth = 3, alpha = 0.6, label = "Theta of a Put Option")
ax.set_xlabel("Spot")
ax.set_ylabel("Theta")
ax.grid()
ax.legend()
plt.show()

fig, ax = plt.subplots(figsize=(12,7))
rho_call =[greeks_bs(c).loc["Rho","Value"] for c in contract_call]
rho_put = [greeks_bs(c).loc["Rho","Value"] for c in contract_put]
ax.plot(spot_vect, rho_call, color = "blue", linewidth = 3, linestyle = "--", label = "Rho of a Call Option")
ax.plot(spot_vect, rho_put, color = "orange", linewidth = 3, alpha = 0.6, label = "Rho of a Put Option")
ax.set_xlabel("Spot")
ax.set_ylabel("Rho")
ax.grid()
ax.legend()
plt.show()

# ALl in one, easier to read
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
delta_call =[greeks_bs(c).loc["Delta","Value"] for c in contract_call]
delta_put = [greeks_bs(c).loc["Delta","Value"] for c in contract_put]
gamma_call =[greeks_bs(c).loc["Gamma","Value"] for c in contract_call]
gamma_put = [greeks_bs(c).loc["Gamma","Value"] for c in contract_put]
vega_call =[greeks_bs(c).loc["Vega","Value"] for c in contract_call]
vega_put = [greeks_bs(c).loc["Vega","Value"] for c in contract_put]
theta_call =[greeks_bs(c).loc["Theta","Value"] for c in contract_call]
theta_put = [greeks_bs(c).loc["Theta","Value"] for c in contract_put]
rho_call =[greeks_bs(c).loc["Rho","Value"] for c in contract_call]
rho_put = [greeks_bs(c).loc["Rho","Value"] for c in contract_put]
axes[0,0].plot(spot_vect, delta_call, label = "Delta Call", color = "blue")
axes[0,0].plot(spot_vect, delta_put, label = "Delta Put", color = "orange")
axes[0,0].set_title("Delta")
axes[0,0].grid()
axes[0,0].legend()
axes[0,1].plot(spot_vect, gamma_call, label = "Gamma Call", color = "blue")
axes[0,1].plot(spot_vect, gamma_put, label = "Gamma Put", color = "orange", linewidth = 3, alpha = 0.6, linestyle = "--")
axes[0,1].set_title("Gamma")
axes[0,1].grid()
axes[0,1].legend()
axes[0,2].plot(spot_vect, vega_call, label = "Vega Call", color = "blue")
axes[0,2].plot(spot_vect, vega_put, label = "Vega Put", color = "orange", linewidth = 3, alpha = 0.6, linestyle = "--")
axes[0,2].set_title("Vega")
axes[0,2].grid()
axes[0,2].legend()
axes[1,0].plot(spot_vect, theta_call, label = "Theta Call", color = "blue")
axes[1,0].plot(spot_vect, theta_put, label = "Theta Put", color = "orange", linewidth = 3, alpha = 0.6, linestyle = "--")
axes[1,0].set_title("Theta")
axes[1,0].grid()
axes[1,0].legend()
axes[1,1].plot(spot_vect, rho_call, label = "Rho Call", color = "blue")
axes[1,1].plot(spot_vect, rho_put, label = "Rho Put", color = "orange")
axes[1,1].set_title("Rho")
axes[1,1].grid()
axes[1,1].legend()
axes[1,2].set_visible(False)
plt.tight_layout()
plt.show()
```
